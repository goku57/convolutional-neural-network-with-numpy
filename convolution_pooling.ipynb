## Kernels 
Cause we do not learn the filters during the exercise we will need some for your experiments. Some Filters are given and you should create at least two more. Do a small research about 'image processing filters' and pick what you like. Remember that your Kernels need to have the same depth as your input. You may consider this issue during your implementation of the convolution operation. 

```python
# typical edge detection filter
class Kernels:
    def __init__(self):
        self.edge_detector_2d_a = np.array(
            [[0., 1., 0.], [1., -4., 1.], [0., 1., 0.]])
        self.sobel_2d = np.array([[1., 2., 1.], [0., 0., 0], [-1., -2., -1.]])
        self.gauss_2d = self.blur()
        self.identical_kernel = np.array(
            [[1., 1., 1.], [1., 1., 1.], [1., 1., 1.]])
        self.smoother = self.identical_kernel/9
        self.edge_detector_2d_b = np.array(
            [[-1., -1., -1.], [-1., 8., -1.], [-1., -1., -1.]])

    def blur(self):
        gauss_1d = scipy.signal.get_window(('gaussian', 1.), 15)
        gauss_2d = np.outer(gauss_1d, gauss_1d)
        return gauss_2d/gauss_2d.sum()
```

## Convolution
**(5)** 
Create a `Conv` class that implements a (naive) convolution operation on _one_ image at the time. Do not use any module, your goal is to get a better understanding for a 2d-conv operation. If your input has more as one channel apply on each the same conv-operation.

```python
class Conv:
    def __init__(self, kernel, stride=1, padding=True, verbose=None):
        """ 
        Args:
            kernel: a filter for the convulution
            stride: step size with which the kernel slides over the image
            padding: if set zero padding will be applied to keep image dimensions
            verbose: if set additional information will be printed, e.g., input and output dimensions
        """
        #self.image_dim = image_dim
        self.kernel = kernel
        self.stride = stride
        self.padding = padding
        self.verbose = verbose

    def forward(self, image):
        """ Executes a convolution on the given image with init params  

        Args:
            image (ndarray): squared image 

        Returns:    
            ndarray: activation map
        """
        image_shape = image.shape

        # adding padding to keep dimensions if needed
        padding = int(
            (image.shape[0]*(self.stride-1) - self.stride + self.kernel.shape[0])/2)
        if (self.padding == True):
            # if image has a third dimension we need to adjust padding
            if (len(image.shape) > 2):
                pad_width = ((padding, padding), (padding, padding), (0, 0))
            else:
                pad_width = ((padding, padding), (padding, padding))
            # create a image with padding
            image = np.pad(image, pad_width, mode='constant',
                           constant_values=0)

        # creating a empty activation map
        a_map_size = int(
            ((image.shape[0] - self.kernel.shape[0])/self.stride)+1)
        a_map = np.zeros([a_map_size, a_map_size])

        # main function
        r2 = 0
        for r in range(0, image.shape[0]-self.kernel.shape[0], self.stride):
            c2 = 0
            for c in range(0, image.shape[0]-self.kernel.shape[0], self.stride):
                # let a be [[1,2,3],[4,5,6],[7,8,9]]
                # with a[0:2,0:2] you would get [[1,2],[4,5]]
                # which extracts a square out of our matrix
                if len(image.shape) > 2:
                    for i in range(image.shape[-1]-1):
                        a_map[r2, c2] += np.sum(image[r:r+self.kernel.shape[0],
                                                      c:c+self.kernel.shape[0], i:i+1]*self.kernel)
                else:
                    a_map[r2, c2] += np.sum(
                        image[r:r+self.kernel.shape[0], c:c+self.kernel.shape[0]]*self.kernel)
                c2 += 1
            r2 += 1

        # additional informations
        if self.verbose == True:
            print("old image dim: ", image_shape,
                  "\nnew image dim: ", a_map.shape)
            print("kernel size:", self.kernel.shape,
                  ", stride size:", self.stride)
        return a_map
```

### $O$ complexity 
Let $n$ be the amount of pixels on a single image, then my code has a complexity of $O(n^2)$. <br> Reason: I have a ``for``-loop inside a ``for``-loop which each $n/s$ iterations. Because we ignore the constant factor $\frac{1}{s}$, the complexity is $O(n^2)$

## Pooling
A `Pooling` class that implements the pooling operation with different functions (max, sum, mean) on a given image.

```python
class Pooling():
    def __init__(self, pooling_function=None, pooling_size=2, stride=2, verbose=None):
        """
        Args:
            pooling_function: defines the pooling operator 'max' (default), 'mean' or 'sum'
            poolig_size: size of one axis of the squared pooling filter
            stride: step size with which the filter slides over the image
            verbose: if set additional information will be printed, e.g., input and output dimensions
        """
        # self.image_dim = image_dim not needed
        self.pooling_function = pooling_function
        self.pooling_size = pooling_size
        self.stride = stride
        self.verbose = verbose

    def forward(self, image):
        """ Executes pooling on the given image with init params

        Args:
            image (ndarray): squared image

        Returns:
            ndarray: activation map
        """
        # calculates the size of the activation map
        a_map_size = (image.shape[0] - self.pooling_size)/self.stride
        a_map_size += 1
        a_map_size = np.uint16(a_map_size)

        # initialize a empty activation map with sizes
        a_map = np.zeros([a_map_size, a_map_size])

        # main function
        r2 = 0
        for r in range(0, image.shape[0]-self.pooling_size, self.stride):
            c2 = 0
            for c in range(0, image.shape[0]-self.pooling_size, self.stride):
                # let a be [[1,2,3],[4,5,6],[7,8,9]]
                # with a[0:2,0:2] you would get [[1,2],[4,5]]
                # which extracts a square out of our matrix
                if self.pooling_function == 'mean':
                    a_map[r2, c2] = np.mean(
                        image[r:r+self.pooling_size, c:c+self.pooling_size])
                elif self.pooling_function == 'sum':
                    a_map[r2, c2] = np.sum(
                        image[r:r+self.pooling_size, c:c+self.pooling_size])
                else:
                    a_map[r2, c2] = np.max(
                        image[r:r+self.pooling_size, c:c+self.pooling_size])
                c2 += 1
            r2 += 1

        # additional informations
        if self.verbose == True:
            print("old image dim: ", image.shape,
                  "\nnew image dim: ", a_map.shape)
            print("pooling function:", self.pooling_function, ", pooling size:",
                  self.pooling_size, ", stride size:", self.stride)
        return a_map
```
